```python
#As seen the figure above data is not lineary separable, 
#hence we will add quadratic features to make the problem linearly separable
def expand(X):
    """
    Adds quadratic features. 
    This expansion allows your linear model to make non-linear separation.
    
    For each sample (row in matrix), compute an expanded row:
    [feature0, feature1, feature0^2, feature1^2, feature0*feature1, 1]
    
    :param X: matrix of features, shape [n_samples,2]
    :returns: expanded features of shape [n_samples,6]
    """
    X_expanded = np.zeros((X.shape[0], 6))
    
    # TODO:<your code here>
    X_expanded[:, 0] = X[:, 0]
    X_expanded[:, 1] = X[:, 1]
    X_expanded[:, 2] = X[:, 0]*X[:, 0]
    X_expanded[:, 3] = X[:, 1]*X[:, 1]
    X_expanded[:, 4] = X[:, 0]*X[:, 1]
    X_expanded[:, 5] = 1
    
    return X_expanded
```

## Logistic Regression

Z = Xw + b

P(y=1|x, w) = sigmod(Z) = 1/(1+e^-Z)

```python
def probability(X, w, b):
    """
    Given input features and weights
    return predicted probabilities of y==1 given x, P(y=1|x), see description above
    
    :param X: feature matrix X of shape [n_samples,6] (expanded)
    :param w: weight vector w of shape [6] for each of the expanded features
    :param b: bias scaler b
    :returns: an array of predicted probabilities in [0,1] interval.
    """

    # TODO:<your code here>
    Z = np.matmul(X, w) + b
    P = 1/(1+np.exp(-Z))
    return P
```

In logistic regression, cross-entropy loss function given as below:

$$ L(w) =  - {1 \over \ell} \sum_{i=1}^\ell \left[ {y_i \cdot log P(y = 1 \, | \, x_i,w,b) + (1-y_i) \cdot log (1-P(y = 1\, | \, x_i,w,b))}\right] $$



```python
def compute_loss(X, y, w, b):
    """
    Given feature matrix X [n_samples,6], target vector [n_samples] of 1/0,
    and weight vector w [6], bias scaler, compute scalar loss function using formula above.
    """
    # TODO:<your code here>
    #predicted probability vector for input examples, shape [n_samples]
    P = probability(X, w, b)
    #number of samples
    l = X.shape[0]
    
    L = -(np.dot(y, np.log(P))+np.dot((1-y), np.log(1-P)))/l
    
    return L
```

```python
def compute_grad(X, y, w, b):
    """
    Given feature matrix X [n_samples,6], target vector [n_samples] of 1/0,
    and weight vector w [6], scaler bias b, compute vector [6] of derivatives of L over each weights.
    """
    
    # TODO<your code here>
    l = X.shape[0]
    P = probability(X, w, b)
    temp = -(y-P)
    db = np.sum(temp)/l
    temp = temp.reshape((1, l))
    dw = np.matmul(temp, X)/l
    return dw.reshape(dw.shape[1]), db
```

```python
from IPython import display

h = 0.01
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

def visualize(X, y, w, b, history):
    """draws classifier prediction with matplotlib magic"""
    Z = probability(expand(np.c_[xx.ravel(), yy.ravel()]), w, b)
    Z = Z.reshape(xx.shape)
    plt.subplot(1, 2, 1)
    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    
    plt.subplot(1, 2, 2)
    plt.plot(history)
    plt.grid()
    ymin, ymax = plt.ylim()
    plt.ylim(0, ymax)
    display.clear_output(wait=True)
    plt.show()
```

## Mini-batch SGD

Stochastic gradient descent just takes a random example on each iteration, calculates a gradient of the loss on it and makes a step:
$$ w_t = w_{t-1} - \eta \dfrac{1}{m} \sum_{j=1}^m \nabla_w L(w_t, b_t, x_{i_j}, y_{i_j}) $$



```python
np.random.seed(42)

X_expanded = expand(X)
w = np.array([0, 0, 0, 0, 0, 1])
b = 0
eta= 0.1 # learning rate

n_iter = 100
batch_size = 4
loss = np.zeros(n_iter)
plt.figure(figsize=(12, 5))

for i in range(n_iter):
    ind = np.random.choice(X_expanded.shape[0], batch_size)
    loss[i] = compute_loss(X_expanded, y, w, b)
    if i % 10 == 0:
        visualize(X_expanded[ind, :], y[ind], w, b, loss)

    # TODO:<your code here>
    dw, db = compute_grad(X_expanded[ind, :], y[ind], w, b)
    w = w - eta*dw
    b= b - eta*db

visualize(X, y, w, b, loss)
plt.clf()
```

```python
X_test = np.array([[1, 2], [0, 0]])
pred = probability(expand(X_test), w, b)
pred
```

## SGD with momentum

Momentum is a method that helps accelerate SGD in the relevant direction and dampens oscillations as can be seen in image below. It does this by adding a fraction $\alpha$ of the update vector of the past time step to the current update vector.
<br>
<br>

$$ \nu_{wt} = \alpha \nu_{wt-1} + \eta\dfrac{1}{m} \sum_{j=1}^m \nabla_w L(w_t, b_t, x_{i_j}, y_{i_j}) $$
$$ w_t = w_{t-1} - \nu_{wt}$$
$$ \nu_{bt} = \alpha \nu_{bt-1} + \eta\dfrac{1}{m} \sum_{j=1}^m \nabla_b L(w_t, b_t, x_{i_j}, y_{i_j}) $$
$$ b_t = b_{t-1} - \nu_{bt}$$

<br>


![](sgd.png)

```python
np.random.seed(42)

X_expanded = expand(X)
w = np.array([0, 0, 0, 0, 0, 1])
b = 0
eta= 0.1 # learning rate
alpha = 0.9 #momentum

n_iter = 100
batch_size = 4
loss = np.zeros(n_iter)
nu_w = np.zeros_like(w)
nu_b = 0
plt.figure(figsize=(12, 5))

for i in range(n_iter):
    ind = np.random.choice(X_expanded.shape[0], batch_size)
    loss[i] = compute_loss(X_expanded, y, w, b)
    if i % 10 == 0:
        visualize(X_expanded[ind, :], y[ind], w, b, loss)

    # TODO:<your code here>
    dw, db = compute_grad(X_expanded[ind, :], y[ind], w, b)
    nu_w = alpha*nu_w + eta*dw
    nu_b = alpha*nu_b + eta*db
    
    w = w - nu_w
    b = b - nu_b

visualize(X, y, w, b, loss)
plt.clf()
```

```python
X_test = np.array([[1, 2], [0, 0]])
pred = probability(expand(X_test), w, b)
pred
```
